## method 01

```python
def non_repeat_character(str):
    for i in str:
        if str.count(i) == 1:
            return i

start = time.time()
print(non_repeat_character(str1M))
print(time.time() - start)
```

## method 02

```python
def non_repeat_character_method02(str):
    rep_chars_dict = {}
    char_list = []
    
    for c in str:
        if c in rep_chars_dict:
            rep_chars_dict[c] += 1
        else:
            rep_chars_dict[c] = 1
            char_list.append(c)
    
    for c in char_list:
        if rep_chars_dict[c] == 1:
            return c
        
start = time.time()
print(non_repeat_character_method02(str1M))
print(time.time() - start)
```
